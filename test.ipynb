```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime

# Define the ticker list
tickers_list = ['NVDA', 'TSLA']

# Fetch the data
data = yf.download(tickers_list, start=datetime(datetime.now().year, 1, 1), end=datetime.now())

# Print first 5 rows of the data
print(data.head())

# Plot the close prices
data = data['Close']
data.plot()

# Calculate the percentage change
data_pct_change = data.pct_change()

# Plot the percentage change
plt.figure(figsize=(14, 7))
for i in range(len(data_pct_change.columns)):
    plt.plot(data_pct_change.index, data_pct_change[data_pct_change.columns[i]], lw=2, label=data_pct_change.columns[i])

plt.legend(loc='upper right', fontsize=12)
plt.ylabel('Price change %')

plt.show()
```
